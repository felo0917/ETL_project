```python
import pandas as pd
from sqlalchemy import create_engine
from secret import username, password
```

```python
#Extract CSVs into DataFrames
athlete_file = "../Resources/athlete_events.csv"
athlete_data_df = pd.read_csv(athlete_file)
athlete_data_df.head()
```

```python
#Create dataframe from specific columns for table Athlete
athlete_columns = ["ID", "Name", "Sex"]
new_athlete_data_df = athlete_data_df[athlete_columns].copy()

#Rename the columns headers
new_athlete_data_df = new_athlete_data_df.rename(columns={"ID": "id"})

#Clean the data by dropping duplicates and setting index
new_athlete_data_df.drop_duplicates("id", inplace=True)
new_athlete_data_df.set_index("id", inplace=True)

new_athlete_data_df.head()
```

```python
#Create dataframe from specific columns for table Olympic Season
oly_season_columns = ["City", "Year", "Season"]
new_oly_season_data_df = athlete_data_df[oly_season_columns].copy()

#Clean the data by dropping duplicates and setting index
new_oly_season_data_df.drop_duplicates("Year", inplace=True)
new_oly_season_data_df.set_index("Year", inplace=True)

new_oly_season_data_df.head()
```

```python
#Create database connection
connection_string = f'{username}:{password}@localhost:5432/Olympics_db'
engine = create_engine(f'postgresql://{connection_string}')
```
