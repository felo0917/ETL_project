```python
import pandas as pd
from sqlalchemy import create_engine
from secret import username, password
```

```python
#import csv - Athletes

athlete_file = "../Resources/athlete_events.csv"
athlete_data_df = pd.read_csv(athlete_file)
athlete_data_df.head()
```

```python
#import csv - Country

population_file = "../Resources/population_GDP.csv"
population_data_df = pd.read_csv(population_file)
population_data = pd.DataFrame(population_data_df)
population_data.head()
```

```python
#create Athlete_event
athlete_event= athlete_data_df[['Medal']]

athlete_event = athlete_event.rename(columns={'Medal':'medal'})
```

```python
athlete_event = athlete_event.dropna()

athlete_event
```

```python
athlete_event.drop_duplicates('medal', inplace=True)

athlete_event
```

```python
#Create dataframe from specific columns for table Athlete
athlete_columns = athlete_data_df[["Name",'Sex','NOC']]


athletes = athlete_columns.drop_duplicates(subset='Name')

athletes
```

```python
athletes['id'] = ""

athletes['id'] = athletes.index + 1

athletes = athletes[['id','Name','Sex','NOC']]

athletes_rename = athletes.rename(columns={'NOC':'country_code'})

athletes_rename
```

```python
athletes_rename.info()
```

```python
#create country df

country_pd = population_data[['Code','Country']]

country_rename = country_pd.rename(columns={'Code':'country_code', 'Country':'country'})

country_rename['id'] = ""

country_rename['id'] = country_rename['id'].index + 1

country = country_rename[['id','country_code','country']]

country
```

```python
athletes_merge = pd.merge(athletes_rename,country, how='inner', on='country_code')

athletes_merge
```

```python
athletes_rename = athletes_merge.rename(columns={'id_x':'id','Name':'name', 'Sex':'sex', 'id_y':'nationality'})

athletes_table = athletes_rename[['id','name','sex', 'nationality']]

athletes_table
```

```python
#create country_stat

country_stat = population_data[[' Population ','GDP per Capita','Code']]

country_stat_rename = country_stat.rename(columns={' Population ': 'population', 'GDP per Capita':'gdp','Code':'country_code'})

country_stat_rename.set_index('population')
```

```python
country_stat_rename["year"] = "2014"
country_stat_rename["country_id"] = ""
```

```python
country_stats = country_stat_rename[["year","population", "gdp","country_code"]] 

country_stats.set_index('year')


country_stats['population'] = country_stats['population'].str.strip()

country_stats
```

```python
country_stats['population'] = country_stats['population'].str.replace(",","")

country_stats
```

```python
coutry_stat_merge = pd.merge(country_stats,country, how='inner', on='country_code')

country_stats_merge_rename = coutry_stat_merge.rename(columns={'id':'country_id'})

```

```python
# country stats table

country_stats_table = country_stats_merge_rename[['year','country_id','population','gdp']]

country_stats_table
```

```python
country_stats_table['population'] = country_stats_table['population'].str.replace(".00","")

country_stats_update = country_stats_table.fillna(0)

country_stats_update
```

```python
#event/athlete_event

events_df = athlete_data_df[['Name','NOC','Games','Year','Season','City','Sport','Event','Medal']]

events_rename = events_df.rename(columns={'Name':'name', 'Sport':'sport'})
```

```python

athletes_id_merge = pd.merge(events_rename,athletes_rename, how='inner', on='name')

athletes_id_merge = athletes_id_merge.rename(columns={'id':'athlete_id'})

athletes_id_merge
```

```python
sport_id = athletes_id_merge[['sport']]

sport_drop = sport_id.drop_duplicates('sport')


```

```python
sport_drop['id'] = ""

sport_reset_index = sport_drop.reset_index()
```

```python
sport_reset_index['id'] = sport_reset_index['id'].index + 1


sport_table = sport_reset_index[['id','sport']]

sport_id_rename = sport_table.rename(columns={'id':'sport_id'})


sport_id_rename
```

```python
sport_athlete_merge = pd.merge(athletes_id_merge,sport_id_rename, how='inner', on='sport')

sport_athlete_merge 
```

```python
#Create dataframe from specific columns for table Olympic Season
oly_season_columns = sport_athlete_merge[["Year","City", "Season"]]

oly_season_columns
```

```python
#Clean the data by dropping duplicates and setting index
oly_season_columns.drop_duplicates("Year", inplace=True)
oly_season_columns.head()
```

```python
oly_season_columns['olympics_season_id'] = ""

oly_season_columns.head()
```

```python
oly_season_rearr = oly_season_columns[['olympics_season_id','Year','City','Season']]

oly_season_rearr.head()
```

```python
oly_season_rearr_reset = oly_season_rearr.reset_index()
oly_season_rearr_column = oly_season_rearr_reset[['olympics_season_id','Year','City','Season']]

oly_season_rearr_column.head()
```

```python
oly_season_rearr_column["olympics_season_id"]= oly_season_rearr_column.index + 1

oly_season_rearr_column.head()
```

```python
olym_sport_athlete_merge = pd.merge(sport_athlete_merge ,oly_season_rearr_column, how='inner', on=['Year','Season','City'])

olym_sport_athlete_merge
```

```python
event_table = olym_sport_athlete_merge[['Event','athlete_id','sport_id','olympics_season_id']]

event_table['id'] = ""

event_table_rearr = event_table[['id','Event','athlete_id','sport_id','olympics_season_id']]

event_rename = event_table_rearr.rename(columns={'Event':'event_name','olympics_season_id':'olympic_season_id'})
```

```python
event_rename["id"]= event_rename.index + 1

event_rename
```

```python
olympic_season = oly_season_rearr_column.rename(columns={'olympics_season_id':'id','Year':'year','City':'city','Season':'season'})
olympic_season.head()
```

```python
event_change = sport_athlete_merge.rename(columns={'Event':'event_name'})

event_merge = pd.merge(event_change,event_rename, how='inner', on=['event_name','athlete_id'])
                                          
athlete_table = event_merge[['athlete_id','Medal','id']]

athlete_drop = athlete_table.dropna()

athlete_table_final = athlete_drop.rename(columns={'Medal':'medal','id':'event_id'})

athlete_table_final
```
